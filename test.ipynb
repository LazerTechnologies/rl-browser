```python
from browser import WebScraper
from pydantic import BaseModel
```

```python
task = "Find the oldest open issue on the PyTorch repository."

class IssueModel(BaseModel):
    date: str
    title: str
    author: str
    description: str

scraper = WebScraper(task, "https://github.com/pytorch/pytorch", IssueModel, callback=print)
await scraper.run()
```

```python
task = """
Add these products to cart and checkout on this store:
- red hoodie

all under $100.
Sizes are M for clothes and 8 for shoes. Men.
"""

class OutputModel(BaseModel):
    total_price: float

scraper = WebScraper(task, "https://demostore.mock.shop/", OutputModel, callback=print)
await scraper.run()
```
